```python
import pandas as pd

data={
    "name":["ritik","ritesh","tanisha"],
    "age":[25,24,"twenty"]
}
df=pd.DataFrame(data)
print(df)
```

```python
df["age"]=pd.to_numeric(df["age"],errors="coerce")
```

```python
print(df)
```

```python
df=df.dropna(subset=["age"])
print(df)
```
